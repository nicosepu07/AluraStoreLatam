### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

tiendas.head()
```

#1. Análisis de facturación



```python
# Crear columna de facturación
tiendas["Facturación"] = tiendas["Precio"] + tiendas["Costo de envío"]

# Facturación total
facturacion_total = tiendas["Facturación"].sum()
facturacion_total
print(f"Facturación total: ${facturacion_total:,.0f}")

```

# 2. Ventas por categoría

```python
# Calcular cantidad de ventas por categoría de producto
ventas_por_categoria = tiendas["Categoría del Producto"].value_counts()

# Mostrar resultados
print("Ventas por categoría de producto:")
print(ventas_por_categoria)
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio por tienda
calificacion_por_tienda = tiendas.groupby("Lugar de Compra")["Calificación"].mean()

print("Calificación promedio por tienda:")
print(calificacion_por_tienda)
```

# 4. Productos más y menos vendidos

```python
# Contar cantidad de ventas por producto
productos_vendidos = tiendas["Producto"].value_counts()

# Identificar productos más y menos vendidos
producto_mas_vendido = productos_vendidos.idxmax()
producto_menos_vendido = productos_vendidos.idxmin()

# Mostrar resultados
print(f"Producto más vendido: {producto_mas_vendido}")
print(f"Producto menos vendido: {producto_menos_vendido}")
```

# 5. Envío promedio por tienda

```python
# Calificación promedio por tienda
calificacion_por_tienda = tiendas.groupby("Lugar de Compra")["Calificación"].mean()

print("Calificación promedio por tienda:")
print(calificacion_por_tienda.round(2))
```
